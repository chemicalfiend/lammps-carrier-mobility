```python
# stole these functions from PyCCE's parser

def yield_index(word, lines, start=0, case_sensitive=False):
    """
    Generator which yields indexes of the lines containing specific word.

    Args:
        word (str): Word to find in the line.
        lines (list of str): List of strings contraining lines from the file. Output of open(file).readlines().
        start (int): First index from which to start search.
        case_sensitive (bool): If True looks for the exact match. Otherwise the search is case insensitive.

    Yields:
        i (int): Index of the line containing word.
    """

    if not case_sensitive:
        word = word.lower()

    for i in range(start, len(lines)):
        lin = lines[i]
        if not case_sensitive:
            lin = lin.lower()

        if word in lin:
            yield i

def find_first_index(word, lines, start=0, case_sensitive=False):
    try:
        ind = next(yield_index(word, lines, start=start, case_sensitive=case_sensitive))
    except StopIteration:
        raise KeyError(f'Could not find {word}')
    return ind
```

# Extracting BBL molecules from data file

```python
#cstart = 94 # line number where coords start (-1)
#cend = 144894 # line number where coords end

filepath = "../BBL-K/5/final.data"


fd = open(filepath, "r")
fo = open("K5.xyz", "w")
f2 = open("../stuff.xyz", "r")


f2lines = f2.readlines()
lines = fd.readlines()

cstart = find_first_index("Atoms", lines, 0, True) + 2
cend = find_first_index("Velocities", lines, 0, True) - 2


mstart = find_first_index("Masses", lines, 0, True) + 2
#print(lines[mstart].split()[0])
mend = cstart - 4


# Setting box bounds

box_bounds = []
for i, line in enumerate(lines):
    line = line.strip()
    
    # Skip empty lines and comments
    if not line or line.startswith('#'):
        continue
        
    # Find box bounds
    if any(s in line for s in ["xlo xhi", "ylo yhi", "zlo zhi"]):
        bounds = list(map(float, line.split()[:2]))
        box_bounds.append(bounds)


# Building a dictionary of atomic indices

ele_dict = {}
for i in range(mstart, mend+1):
    line = lines[i]
    tokens = line.split()
    index = int(tokens[0])
    element = ele.element_from_mass(float(tokens[1]))
    ele_dict[index] = element    


mnos = []
for i in range(cstart, cend):
    line = lines[i]
    tokens = line.split()
    mol = int(tokens[1])
    mnos.append(mol)

num_mols = max(mnos)

mols = [[] for i in range(num_mols)]

atoms = []

cou = 2
for i in range(cstart, cend):
    line = lines[i]
    tokens = line.split()
    n = int(tokens[0])
    mol = int(tokens[1])
    typeid = int(tokens[2])
    atomtype = ele_dict[typeid].symbol

    charge = float(tokens[3])
    x = float(tokens[4])
    y = float(tokens[5])
    z = float(tokens[6])
    tok2 = f2lines[cou].split()
    x2 = float(tok2[0])
    y2 = float(tok2[1])
    z2 = float(tok2[2])

    atom = [n, mol, atomtype, charge, x2, y2, z2]
    atoms.append(atom)
    mols[mol-1].append(atom)


# TODO : From atoms in the same mol, find an atom that's only in BBL such that everything other than BBL is excluded.

BBL_mols = []
BBL_atoms = []

for atom in atoms:
    ats = [atom[2] for atom in mol]
    #print(ats)
    flag = False
    if 'N' in ats and 'O' in ats and 'C' in ats:
        flag = True
    if flag:
        BBL_mols.append(mol)
        for atom in mol:
            BBL_atoms.append(atom)

natoms = len(BBL_atoms)

fo.write(f"{natoms}\n")
fo.write("Generated by LammpsCT\n")

#for mol in BBL_mols:
#    for a in mol:
for a in BBL_atoms:
    fo.write(f"{a[2]} {a[4]} {a[5]} {a[6]} \n")

#for a in atoms:
#    fo.write(f"{a[0]} {a[1]} {a[2]} {a[3]} {a[4]} {a[5]} {a[6]} \n")

fo.close()
fd.close()
```

```python
# Experimental unwrapping code
BBL_atoms.sort(key = lambda a: a[0])
coords = [np.array([a[4], a[5], a[6]]) for a in BBL_atoms]
unwrapped = copy.deepcopy(BBL_atoms)

for i in range(1, len(BBL_atoms)):    
    delta = coords[i] - coords[i-1]
    
    # Adjust for periodic boundary crossings
    # If displacement is greater than half the box, it's likely a boundary crossing
    for dim in range(3):
        if delta[dim] > box_bounds[dim][0] / 2:
            delta[dim] -= box_bounds[dim][0]
        elif delta[dim] < -box_bounds[dim][1] / 2:
            delta[dim] += box_bounds[dim][1]
    
    # Calculate new unwrapped position
    unwrapped[i][4] = unwrapped[i-1][4] + delta[0]
    unwrapped[i][5] = unwrapped[i-1][5] + delta[1]
    unwrapped[i][6] = unwrapped[i-1][6] + delta[2]
```
